```python
import pandas as pd
import numpy as np

nhtsa_headers = [
    "CMPLID", "ODINO", "MFR_NAME", "MAKETXT", "MODELTXT", "YEARTXT", "CRASH",
    "FAILDATE", "FIRE", "INJURED", "DEATHS", "COMPDESC", "CITY", "STATE",
    "VIN", "DATEA", "LDATE", "MILES", "OCCURENCES", "CDESCR", "CMPL_TYPE",
    "POLICE_RPT_YN", "PURCH_DT", "ORIG_OWNER_YN", "ANTI_BRAKES_YN",
    "CRUISE_CONT_YN", "NUM_CYLS", "DRIVE_TRAIN", "FUEL_SYS", "FUEL_TYPE",
    "TRANS_TYPE", "VEH_SPEED", "DOT", "TIRE_SIZE", "LOC_OF_TIRE",
    "TIRE_FAIL_TYPE", "ORIG_EQUIP_YN", "MANUF_DT", "SEAT_TYPE",
    "RESTRAINT_TYPE", "DEALER_NAME", "DEALER_TEL", "DEALER_CITY",
    "DEALER_STATE", "DEALER_ZIP", "PROD_TYPE", "REPAIRED_YN",
    "MEDICAL_ATTN", "VEHICLES_TOWED_YN", "STATE_OF_INCIDENT", "VEHICLE_OPERATOR"
]

complaint_data_10_14 = pd.read_csv('Complaint_2010_2014.csv', header = None, names = nhtsa_headers)
complaint_data_15_19 = pd.read_csv('Complaint_2015_2019.csv', header = None, names = nhtsa_headers)
complaint_data_20_24 = pd.read_csv('Complaint_2020_2024.csv', header = None, names = nhtsa_headers)
complaint_data_25_26 = pd.read_csv('Complaint_2025_2026.csv', header = None, names = nhtsa_headers)

complaint_data_combined = pd.concat([complaint_data_10_14, complaint_data_15_19, complaint_data_20_24, complaint_data_25_26])
```

```python
recall_data_10_14 = pd.read_csv('Recall_2010_2014.csv')
recall_data_15_19 = pd.read_csv('Recall_2015_2019.csv')
recall_data_20_24 = pd.read_csv('Recall_2020_2024.csv')
recall_data_25_26 = pd.read_csv('Recall_2025_2026.csv')

recall_data_combined = pd.concat([recall_data_10_14, recall_data_15_19, recall_data_20_24, recall_data_25_26])
```

```python
complaint_data_combined["YEARTXT"] = pd.to_numeric(complaint_data_combined["YEARTXT"], errors="coerce")
complaint_data_combined = complaint_data_combined.dropna(subset=["YEARTXT"])
complaint_data_combined["YEARTXT"] = complaint_data_combined["YEARTXT"].astype(int)
complaint_data_combined = complaint_data_combined.loc[complaint_data_combined['YEARTXT'] != 9999]
```

```python
PASSENGER_MAKES = {
    "ACURA", "ALFA ROMEO", "ASTON MARTIN", "AUDI", "BENTLEY", "BMW",
    "BUICK", "CADILLAC", "CHEVROLET", "CHRYSLER", "DODGE", "FERRARI",
    "FIAT", "FORD", "GEO", "GMC", "HONDA", "HUMMER", "HYUNDAI",
    "INFINITI", "JAGUAR", "JEEP", "KIA", "LAMBORGHINI", "LAND ROVER",
    "LEXUS", "LINCOLN", "LOTUS", "LUCID", "MASERATI", "MAZDA",
    "MCLAREN", "MERCEDES-BENZ", "MERCEDES", "MERCURY", "MINI",
    "MITSUBISHI", "NISSAN", "OLDSMOBILE", "PLYMOUTH", "POLESTAR",
    "PONTIAC", "PORSCHE", "RAM", "RIVIAN", "ROLLS-ROYCE", "SAAB",
    "SATURN", "SCION", "SMART", "SUBARU", "TESLA", "TOYOTA",
    "VINFAST", "VOLKSWAGEN", "VOLVO"
}

complaint_data_combined = complaint_data_combined[complaint_data_combined["MAKETXT"].str.upper().isin(PASSENGER_MAKES)]

for col in ["CRASH", "FIRE"]:
    complaint_data_combined[col] = complaint_data_combined[col].map({"Y": 1, "N": 0}).fillna(0).astype(int)

complaint_data_combined['FAILDATE'] = pd.to_datetime(complaint_data_combined['FAILDATE'].astype(str), format='%Y%m%d', errors='coerce')

complaint_data_combined_year = complaint_data_combined[(complaint_data_combined["YEARTXT"] >= 2010) & (complaint_data_combined["YEARTXT"] <= 2021)]

complaint_data_combined_year = complaint_data_combined_year.groupby(["MAKETXT", "MODELTXT", "YEARTXT"]).agg(
    complaint_count = ("ODINO", "count"),
    crash_count = ("CRASH", "sum"),
    fire_count = ("FIRE", "sum"),  
    injured_count = ("INJURED", "sum"),
    death_count = ("DEATHS", "sum"),
    first_complaint_date = ("FAILDATE", "min"),
    last_complaint_date = ("FAILDATE", "max"),
    description = ("COMPDESC", lambda x: " ".join(x.dropna()))).reset_index()

complaint_data_combined_year['crash_ratio'] = complaint_data_combined_year['crash_count'] / complaint_data_combined_year['complaint_count']
complaint_data_combined_year['fire_ratio'] = complaint_data_combined_year['fire_count'] / complaint_data_combined_year['complaint_count']
complaint_data_combined_year['injured_ratio'] = complaint_data_combined_year['injured_count'] / complaint_data_combined_year['complaint_count']
complaint_data_combined_year['death_ratio'] = complaint_data_combined_year['death_count'] / complaint_data_combined_year['complaint_count']
complaint_data_combined_year["days_span"] = (complaint_data_combined_year["last_complaint_date"] - complaint_data_combined_year["first_complaint_date"]).dt.days
complaint_data_combined_year["complaints_per_day"] = complaint_data_combined_year["complaint_count"] / complaint_data_combined_year["days_span"]
complaint_data_combined_year = complaint_data_combined_year.replace([np.inf, -np.inf], 0)

complaint_data_combined_year = complaint_data_combined_year.rename(columns = {
    "MAKETXT": "MAKE",
    "MODELTXT": "MODEL",
    "YEARTXT": "YEAR"})
```

```python
PASSENGER_MAKES = {
    "ACURA", "ALFA ROMEO", "ASTON MARTIN", "AUDI", "BENTLEY", "BMW",
    "BUICK", "CADILLAC", "CHEVROLET", "CHRYSLER", "DODGE", "FERRARI",
    "FIAT", "FORD", "GEO", "GMC", "HONDA", "HUMMER", "HYUNDAI",
    "INFINITI", "JAGUAR", "JEEP", "KIA", "LAMBORGHINI", "LAND ROVER",
    "LEXUS", "LINCOLN", "LOTUS", "LUCID", "MASERATI", "MAZDA",
    "MCLAREN", "MERCEDES-BENZ", "MERCEDES", "MERCURY", "MINI",
    "MITSUBISHI", "NISSAN", "OLDSMOBILE", "PLYMOUTH", "POLESTAR",
    "PONTIAC", "PORSCHE", "RAM", "RIVIAN", "ROLLS-ROYCE", "SAAB",
    "SATURN", "SCION", "SMART", "SUBARU", "TESLA", "TOYOTA",
    "VINFAST", "VOLKSWAGEN", "VOLVO"
}

recall_data_combined = recall_data_combined[recall_data_combined["MAKE"].str.upper().isin(PASSENGER_MAKES)]
recall_data_combined = recall_data_combined.drop_duplicates(subset = ["MAKE", "MODEL", "MODEL YEAR"], keep='last')
recall_data_combined_year = recall_data_combined[(recall_data_combined["MODEL YEAR"] >= 2010) & (recall_data_combined["MODEL YEAR"] <= 2021)]
recall_data_combined_year['Recall'] = 1

recall_data_combined_year = recall_data_combined_year.rename(columns = {"MODEL YEAR": "YEAR"})
```

```python
complaint_recall = complaint_data_combined_year.merge(recall_data_combined_year[['MAKE', 'MODEL', 'YEAR', 'Recall']], on = ['MAKE', 'MODEL', 'YEAR'], how = 'left')
complaint_recall                                                 
```

```python
complaint_recall["Recall"] = pd.to_numeric(complaint_recall["Recall"], errors="coerce")
complaint_recall["Recall"] = complaint_recall["Recall"].fillna(0)
complaint_recall["Recall"] = complaint_recall["Recall"].astype(int)
```

```python
complaint_recall["Recall"].value_counts()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
import pandas as pd
import re

final_df = complaint_recall

def clean_text(text):
    if not isinstance(text, str):
        return ""
    text = text.lower()
    text = re.sub(r'[^a-z\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

final_df["summary_clean"] = final_df["description"].apply(clean_text)

# fit TF-IDF on complaint summaries
tfidf = TfidfVectorizer(
    max_features=100,    # top 100 words
    ngram_range=(1, 2),  # single words and two-word phrases
    stop_words="english",
    min_df=5             # word must appear in at least 5 vehicles
)

X_tfidf = tfidf.fit_transform(final_df["summary_clean"])

# quick logistic regression to find which words predict recalls
lr = LogisticRegression(class_weight="balanced")
lr.fit(X_tfidf, final_df["Recall"])

# see top words associated with recall
feature_names = tfidf.get_feature_names_out()
coefficients = lr.coef_[0]

word_importance = pd.DataFrame({
    "word": feature_names,
    "coefficient": coefficients
}).sort_values("coefficient", ascending=False)

print("Top 20 words predicting RECALL:")
print(word_importance.head(20))

print("\nTop 20 words predicting NO RECALL:")
print(word_importance.tail(20))
```

```python
RECALL_WORDS = [
    # from TF-IDF
    "fuel",
    "engine",
    "seat",
    "seats",
    "electrical",
    "structure",
    "exterior lighting",
    "brakes",
    "power train",
    "cooling",
    "airbag",
    "air bag",
    "steering",
    
    # domain knowledge additions
    "stalled",
    "stall",
    "fire",
    "smoke",
    "explod",        # catches exploded/exploding
    "accelerat",     # catches accelerating/acceleration
    "rollover",
    "lost control",
    "without warning",
    "sudden",
    "failed",
    "defect",
    "caught fire",
    "would not stop",
    "brake failure",
    "gas leak",
    "fuel leak",
    "transmission",
]

NON_RECALL_WORDS = [
    "control", "brakes engine", "electrical steering",
    "train electrical", "propulsion", "engine unknown",
    "engine steering", "automatic", "stability control"
]

def recall_language_score(text):
    if not text:
        return 0
    words = text.split()
    return sum(text.count(w) for w in RECALL_WORDS) / (len(words) + 1)

def safe_language_score(text):
    if not text:
        return 0
    words = text.split()
    return sum(text.count(w) for w in NON_RECALL_WORDS) / (len(words) + 1)

final_df["recall_language_score"] = final_df["summary_clean"].apply(recall_language_score)
final_df["safe_language_score"] = final_df["summary_clean"].apply(safe_language_score)

# sanity check — recalled vehicles should have higher recall_language_score
print(final_df.groupby("Recall")[["recall_language_score", "safe_language_score"]].mean())
```

```python
RECALL_WORDS_WEIGHTED = {
    "caught fire": 5,
    "explod": 5,
    "rollover": 5,
    "lost control": 5,
    "brake failure": 5,
    "would not stop": 5,
    "fuel leak": 4,
    "gas leak": 4,
    "stall": 4,
    "stalled": 4,
    "without warning": 4,
    "accelerat": 4,
    "smoke": 3,
    "sudden": 3,
    "airbag": 3,
    "air bag": 3,
    "defect": 3,
    "failed": 2,
    "death": 5,
    "injury": 4,
    "injured": 4,
}

def recall_language_score(text):
    if not text:
        return 0
    words = text.split()
    score = sum(
        weight * text.count(word)
        for word, weight in RECALL_WORDS_WEIGHTED.items()
    )
    return score / (len(words) + 1)

final_df["recall_language_score"] = final_df["summary_clean"].apply(recall_language_score)

# sanity check
print(final_df.groupby("Recall")[["recall_language_score"]].mean())
```

```python
manufacturer_stats = final_df.groupby("MAKE").agg(
    total_models=("Recall", "count"),
    total_recalled=("Recall", "sum")
).reset_index()

manufacturer_stats["manufacturer_recall_rate"] = (
    manufacturer_stats["total_recalled"] / manufacturer_stats["total_models"]
)

print(manufacturer_stats.sort_values("manufacturer_recall_rate", ascending=False))
```

```python
import xgboost as xgb
from xgboost import XGBClassifier, plot_importance
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_validate, learning_curve
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, mean_squared_error, mean_absolute_error, r2_score, median_absolute_error, PrecisionRecallDisplay, make_scorer
import matplotlib.pyplot as plt


x_var = ['crash_ratio', 'fire_ratio', 'injured_ratio', 'death_ratio', 'complaints_per_day']
y_var = 'Recall'

X = complaint_recall[x_var]
y = complaint_recall[y_var]

train_x, test_x, train_y, test_y = train_test_split(X, y, test_size = 0.2, stratify = y)
xgb_model = xgb.XGBClassifier(n_estimators=100, eval_metric='logloss')
xgb_model.fit(train_x, train_y)
```

```python
skf = StratifiedKFold(n_splits=10)

scoring = {
    'accuracy': 'accuracy',
    'precision': make_scorer(precision_score, zero_division=0),
    'recall': make_scorer(recall_score, zero_division=0),
    'f1': make_scorer(f1_score, zero_division=0)
}

scores = cross_validate(xgb_model, X, y, cv = skf, scoring = scoring)

df_scores = pd.DataFrame.from_dict(scores)

df_scores.mean()
```

```python
cv = StratifiedKFold(n_splits=10)

train_sizes = np.linspace(0.1, 1.0, 10)
    
train_sizes, train_scores, valid_scores = learning_curve(
    xgb_model, X, y, 
    cv=cv,
    n_jobs=-1,
    train_sizes=train_sizes,
    scoring='neg_log_loss'
)

train_mean = -np.mean(train_scores, axis=1)
train_std = -np.std(train_scores, axis=1)
valid_mean = -np.mean(valid_scores, axis=1)
valid_std = -np.std(valid_scores, axis=1)

plt.figure(figsize=(10, 6))
plt.plot(train_sizes, train_mean, label='Training score', color='blue')
plt.plot(train_sizes, valid_mean, label='Cross-validation score', color='red')

plt.title(f'Learning Curves')
plt.xlabel('Training Examples')
plt.ylabel('Log Loss')
plt.grid(True)
plt.legend(loc='lower right')
plt.tight_layout()
plt.show()
```
